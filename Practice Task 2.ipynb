```python
name_of_file= 'kinopoisk.csv'
path = '/content/drive/My Drive/Colab Notebooks/'+ name_of_file
```

```python
data_ab = pd.read_csv(path, sep=',', decimal='.')
```

```python
columns_mapping = {
    'film-item-rating-position__position':'Позиция в рейтинге',
    'selection-film-item-poster__rating':'Рейтинг 1',
    'selection-film-item-meta__name':'Название на русском',
    'rating__value':'Рейтинг 2',
    'rating__count':'Количество оценок',
    'selection-film-item-meta__meta-additional-item':'Страна',
    'selection-film-item-meta__meta-additional-item 2':'Жанр'
}
```

```python
data_ab = data_ab.drop(['HaRNLxUnsJBznTziWnJLc', 'HaRNLxUnsJBznTziWnJLc href', 'selection-film-item-meta__link href', 'selection-film-item-meta__original-name'], axis='columns')
```

```python
data_ab = data_ab.rename(columns=columns_mapping)
```

```python
data_ab
```

H0 - рейтинги не отличаются

```python
result=mannwhitneyu(data_ab['Рейтинг 1'],data_ab['Рейтинг 2'])
```

```python
print(f'P-value для нулевой гипотезы H0 имеет значение {result.pvalue}')
```

Основываясь на значении P-value, можно сделать вывод о том, что при использовании критерия Манна-Уитни нулевая гипотеза H0 верна, так как P-value > 0.05. Значит, старый и новый рейтинги фильмов из "Топ 250 фильмов" сайта "Кинопоиск" не отличаются друг от друга.
